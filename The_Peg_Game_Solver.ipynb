### Run this game solver to find all possible solutions

```python
def make_move_up_left(i,j,tokens):
          
    if ((i+2>=0 and i+2<=4 and j-2>=0 and j-2<=2-i) and ((i,j) in tokens) and ((i+1,j-1) in tokens) and (i+2,j-2) not in tokens):
        val=1
        tokens_new=copy.deepcopy(tokens)
        tokens_new.remove((i,j))
        tokens_new.remove((i+1,j-1))
        tokens_new.append((i+2,j-2))
        return tokens_new,val
    else:
        val=0
        return tokens,val

def make_move_up_right(i,j,tokens):
     
    if ((i+2>=0 and i+2<=4 and j>=0 and j<=2-i) and ((i,j) in tokens) and ((i+1,j) in tokens) and (i+2,j) not in tokens):
        val=1
        tokens_new=copy.deepcopy(tokens)
        tokens_new.remove((i,j))
        tokens_new.remove((i+1,j))
        tokens_new.append((i+2,j))
        return tokens_new,val
    else:
        val=0
        return tokens,val    
    
def make_move_down_left(i,j,tokens):
     
    if ((i-2>=0 and i-2<=4 and j>=0 and j<=6-i) and ((i,j) in tokens) and ((i-1,j) in tokens) and (i-2,j) not in tokens):
        val=1
        tokens_new=copy.deepcopy(tokens)
        tokens_new.remove((i,j))
        tokens_new.remove((i-1,j))
        tokens_new.append((i-2,j))
        return tokens_new,val
    else:
        val=0
        return tokens,val
    
def make_move_down_right(i,j,tokens):
     
    if ((i-2>=0 and i-2<=4 and j+2>=0 and j+2<=6-i) and ((i,j) in tokens) and ((i-1,j+1) in tokens) and (i-2,j+2) not in tokens):
        val=1
        tokens_new=copy.deepcopy(tokens)
        tokens_new.remove((i,j))
        tokens_new.remove((i-1,j+1))
        tokens_new.append((i-2,j+2))
        return tokens_new,val
    else:
        val=0
        return tokens,val
    
def make_move_left(i,j,tokens):
     
    if ((i>=0 and i<=4 and j-2>=0 and j-2<=4-i) and ((i,j-1) in tokens) and (i,j-2) not in tokens):
        val=1
        tokens_new=copy.deepcopy(tokens)
        tokens_new.remove((i,j))
        tokens_new.remove((i,j-1))
        tokens_new.append((i,j-2))
        return tokens_new,val
    else:
        val=0
        return tokens,val
    
def make_move_right(i,j,tokens):
     
    if ((i>=0 and i<=4 and j+2>=0 and j+2<=4-i) and ((i,j+1) in tokens) and (i,j+2) not in tokens):
        val=1
        tokens_new=copy.deepcopy(tokens)
        tokens_new.remove((i,j))
        tokens_new.remove((i,j+1))
        tokens_new.append((i,j+2))
        return tokens_new,val
    else:
        val=0
        return tokens,val   

    
def make_ind_move(i,j,tokens,mode):
    if (i,j) not in tokens:
        return tokens,None
    else:
        if mode==0:
            tokens_new,val=make_move_up_left(i,j,tokens)
            if val!=0:
                move=(i,j,mode)
                return tokens_new,move
            else:
                return tokens,None
        elif mode==1:
            tokens_new,val=make_move_up_right(i,j,tokens)
            if val!=0:
                move=(i,j,mode)
                return tokens_new,move
            else:
                return tokens,None
        elif mode==2:
            tokens_new,val=make_move_down_left(i,j,tokens)
            if val!=0:
                move=(i,j,mode)
                return tokens_new,move
            else:
                return tokens,None
        elif mode==3:
            tokens_new,val=make_move_down_right(i,j,tokens)
            if val!=0:
                move=(i,j,mode)
                return tokens_new,move
            else:
                return tokens,None
        elif mode==4:
            tokens_new,val=make_move_left(i,j,tokens)
            if val!=0:
                move=(i,j,mode)
                return tokens_new,move
            else:
                return tokens,None
        elif mode==5:
            tokens_new,val=make_move_right(i,j,tokens)
            if val!=0:
                move=(i,j,mode)
                return tokens_new,move
            else:
                return tokens,None
        else:
            raise ValueError("Moving Mode Violation")

def converting_moves(path):
    
    converted_path=[]
    for move in path:
        if move[2]==0:
            converted_move=((move[0],move[1]),(move[0]+1,move[1]-1),(move[0]+2,move[1]-2))
        elif move[2]==1:
            converted_move=((move[0],move[1]),(move[0]+1,move[1]),(move[0]+2,move[1]))
        elif move[2]==2:
            converted_move=((move[0],move[1]),(move[0]-1,move[1]),(move[0]-2,move[1]))
        elif move[2]==3:
            converted_move=((move[0],move[1]),(move[0]-1,move[1]+1),(move[0]-2,move[1]+2))
        elif move[2]==4:
            converted_move=((move[0],move[1]),(move[0],move[1]-1),(move[0],move[1]-2))
        elif move[2]==5:
            converted_move=((move[0],move[1]),(move[0],move[1]+1),(move[0],move[1]+2))
        converted_path.append(converted_move)
    return converted_path




    
def Triagle_BackTrack():
    global tokens_path,path,solutions
    current_tokens=tokens_path[-1]
      
    for token in current_tokens:        
        for mode in range(6):
            new_tokens,move=make_ind_move(token[0],token[1],current_tokens,mode)
            if bool(move):
                tokens_path.append(new_tokens)
                path.append(move)
                Triagle_BackTrack()
    if len(path)==13:
        solutions.append(copy.deepcopy(path))
    if path!=[]:
        tokens_path.pop()
        path.pop()
    return

solutions=[]
path=[]
tokens=[]
tokens_path=[]
for i in range(4):
    for j in range(5-i):
        tokens.append((i,j))
tokens_path.append(tokens)
Triagle_BackTrack()

copy_solutions=copy.deepcopy(solutions)
unique_solutions_temp=[]
unique_solutions=[]
unique_inds=[]
for solution in copy_solutions:
    if sorted(solution) not in unique_solutions_temp:
        unique_inds.append(solutions.index(solution))
        unique_solutions_temp.append(sorted(solution))
for unique_ind in unique_inds:
    unique_solutions.append(solutions[unique_ind])
print("There are {} unique solutions among {} possible solutions".format(len(unique_solutions),len(solutions)))    
```

### Run the code below to see animation of the selected solution (If getting error, run the code twice)

```python
def create_pencil(color,width,speed=10):
    pencil = turtle.Turtle(visible=False)
    pencil.color('black')
    pencil.pensize(width=6)
    pencil.speed(speed)
    return pencil
def board_draw(pencil):
    pencil.penup()
    pencil.color('black')
    pencil.setposition(-300,-200) 
    pencil.fillcolor(0.8,0.6,0.2)
    pencil.begin_fill()
    pencil.pendown()
    pencil.forward(600)
    pencil.left(120)
    pencil.forward(600)
    pencil.left(120)
    pencil.forward(600)
    pencil.left(120)
    pencil.end_fill()
def circle_draw(pencil,x,y,color,R=30):
    pencil.penup()
    pencil.color('black')
    pencil.setposition(x,y)
    pencil.fillcolor(color)
    pencil.begin_fill()
    pencil.pendown()
    pencil.circle(R)
    pencil.end_fill()
    
class Token():
    def __init__(self,ind1,ind2,color,value=0):
        self.ind1=ind1
        self.ind2=ind2
        self.color=color
        self.value=value
    def update_value(self,new_value):
        self.value=new_value
    def update_color(self,new_color):
        self.color=new_color
        
all_tokens=[]
color_list=['red','yellow','green','blue','white']
valid_inds=[]
for i in range(5):
    tokens=[]
    for j in range(5-i):
        tokens.append(Token(i,j,color_list[i],1))
        valid_inds.append((i,j))
    all_tokens.append(tokens)
all_tokens[4][0].update_value(0)
all_tokens[4][0].update_color(color_list[4])

step_h=90
step_diag_v=round(step_h*math.sqrt(3)/2)
step_diag_h=90/2
all_coordinates=[]
all_colors=[]

for i in range(5,0,-1):
    row_coordinates=[]
    colors=[]
    color=color_list[5-i]
    for j in range(i):
        row_coordinates.append((-180+step_diag_h*(5-i)+step_h*j,-155+step_diag_v*(5-i)))
        colors.append(color)
    all_colors.append(colors)
    all_coordinates.append(row_coordinates) 

pencil=create_pencil('black',6,speed=0)

pencil.penup()
pencil.setposition(0,330) 
pencil.color('brown')
style = ('Courier', 40, 'bold')
pencil.write('Tricky Triangle Game!', font=style, align='center')
pencil.hideturtle()

board_draw(pencil)
tscreen = turtle.Screen()
tscreen.bgcolor(0.4,0.75,0.9)

for ind1 in range(len(all_coordinates)):
    for ind2 in range(len(all_coordinates[ind1])):
        color=all_tokens[ind1][ind2].color
        circle_draw(pencil,all_coordinates[ind1][ind2][0],all_coordinates[ind1][ind2][1],color)

import time

time.sleep(3)

pencil.penup()
pencil.setposition(0,-300) 
pencil.color('blue')
style = ('Courier', 32, 'bold')
pencil.write('Playing the solution N={}'.format(ind), font=style, align='center')
pencil.hideturtle()
time.sleep(2)

for move in moves:
    color=all_tokens[move[0][0]][move[0][1]].color
    circle_draw(pencil,all_coordinates[move[2][0]][move[2][1]][0],all_coordinates[move[2][0]][move[2][1]][1],color)
    circle_draw(pencil,all_coordinates[move[1][0]][move[1][1]][0],all_coordinates[move[1][0]][move[1][1]][1],'white')
    circle_draw(pencil,all_coordinates[move[0][0]][move[0][1]][0],all_coordinates[move[0][0]][move[0][1]][1],'white')
    all_tokens[move[2][0]][move[2][1]].update_color(all_tokens[move[0][0]][move[0][1]].color)
    all_tokens[move[1][0]][move[1][1]].update_color('white')
    all_tokens[move[0][0]][move[0][1]].update_color('white')
    time.sleep(1)

tscreen.exitonclick()
```
